```python
"""
This code converts the benchmarking results obtained using 
src/keypoints/benchmark/MP_MEE_sidebyside/MP_MEE_sidebyside.py
to a visual representation.

Benchmark data is expected to contain timestamps for mediapipe,
mediapipe with tracking enabled and the meeshape model.
"""

import pandas as pd
import matplotlib.pyplot as plt

df_bench = pd.read_csv('MP_MEE_sidebyside/MP_MEE_comparison_CompleteFlaccid1.csv')

plt.plot(df_bench['mediapipe'])
plt.plot(df_bench['mediapipe_tracking'])
plt.plot(df_bench['meeshape'])
plt.title('Keypoint detectie benchmark')
plt.ylabel('Time (in ms)')
plt.xlabel('frame nummer')
plt.legend(['Mediapipe', 'Mediapipe (tracking)', 'MEE Shape'], loc='upper left')
plt.savefig('benchmark-all.png')
plt.show()

plt.plot(df_bench[:500]['mediapipe'])
plt.plot(df_bench[:500]['mediapipe_tracking'])
plt.plot(df_bench[:500]['meeshape'])
plt.title('Keypoint detectie benchmark')
plt.ylabel('Time (in ms)')
plt.xlabel('frame nummer')
plt.legend(['Mediapipe', 'Mediapipe (tracking)', 'MEE Shape'], loc='upper left')
plt.savefig('benchmark-500.png')
plt.show()

plt.plot(df_bench[:100]['mediapipe'])
plt.plot(df_bench[:100]['mediapipe_tracking'])
plt.plot(df_bench[:100]['meeshape'])
plt.title('Keypoint detectie benchmark')
plt.ylabel('Time (in ms)')
plt.xlabel('frame nummer')
plt.legend(['Mediapipe', 'Mediapipe (tracking)', 'MEE Shape'], loc='upper left')
plt.savefig('benchmark-100.png')
plt.show()
```
